```python
from transformers import BartTokenizer, BartForConditionalGeneration

def summarize_text(input_text):
    # Load pre-trained BART model and tokenizer
    model_name = "facebook/bart-large-cnn"
    tokenizer = BartTokenizer.from_pretrained(model_name)
    model = BartForConditionalGeneration.from_pretrained(model_name)

    # Tokenize input text
    inputs = tokenizer(input_text, return_tensors="pt", max_length=1024, truncation=True)

    # Generate summary
    summary_ids = model.generate(inputs["input_ids"], max_length=150, length_penalty=2.0, num_beams=4, early_stopping=True)

    # Decode the summary
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

    return summary

# Example usage
input_text = """
Artificial Intelligence (AI) is a branch of computer science that aims to create machines capable of intelligent behavior. It involves the development of algorithms and models that enable computers to perform tasks that typically require human intelligence. These tasks include learning from experience, understanding natural language, recognizing patterns, and making decisions.

AI has applications in various fields, including healthcare, finance, education, and autonomous systems. Machine learning, a subset of AI, focuses on developing systems that can learn and improve from experience. Deep learning, a subfield of machine learning, involves neural networks with multiple layers, inspired by the structure of the human brain.

The field of AI is rapidly evolving, with ongoing research and advancements. As technology continues to progress, AI is expected to play a crucial role in shaping the future of various industries.
"""

summary = summarize_text(input_text)
print("Original Text:")
print(input_text)
print("\nSummarized Text:")
print(summary)
```

```python
from transformers import BartTokenizer, BartForConditionalGeneration

def summarize_large_text(large_text):
    # Load pre-trained BART model and tokenizer
    model_name = "facebook/bart-large-cnn"
    tokenizer = BartTokenizer.from_pretrained(model_name)
    model = BartForConditionalGeneration.from_pretrained(model_name)

    # Tokenize and truncate the large text
    inputs = tokenizer(large_text, return_tensors="pt", max_length=1024, truncation=True)

    # Generate summary
    summary_ids = model.generate(inputs["input_ids"], max_length=150, length_penalty=2.0, num_beams=4, early_stopping=True)

    # Decode the summary
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

    return summary

# Larger Text for Summarization
large_text = """
Reading is a fundamental skill that plays a crucial role in our intellectual and personal development. Beyond its basic function of decoding written words, reading opens up a world of knowledge, imagination, and empathy.

In the realm of education, reading is the foundation of learning. It is through reading that we acquire information, expand our vocabulary, and develop critical thinking skills. Whether it's textbooks, literature, or scientific papers, reading is the gateway to understanding various subjects and disciplines.

Moreover, reading has a profound impact on our cognitive abilities. It stimulates the brain, enhances concentration, and improves memory. Engaging with diverse written materials exposes us to different perspectives and cultures, fostering a broadened worldview.

Beyond the educational context, reading is a source of entertainment and escape. Fictional stories transport us to fantastical realms, while non-fiction provides insights into the lives and experiences of others. It is a tool for relaxation, a means to unwind and detach from the demands of daily life.

In a world inundated with information, the ability to discern, analyze, and synthesize written content is more valuable than ever. Reading equips us with the skills to navigate the complexities of the modern age.

In summary, reading is not merely a skill; it is a multifaceted tool that empowers us intellectually, emotionally, and socially. Embracing the habit of reading is an investment in personal growth and a key to unlocking the vast reservoir of human knowledge.
"""

# Generate and print the summary
summary = summarize_large_text(large_text)
print("Original Text:")
print(large_text)
print("\nSummarized Text:")
print(summary)
```
